```python
import pandas as pd 
import numpy as np
from autots import AutoTS
```

```python
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('C:\\Users\\user\\saurabh sir pandas\\ML\\Auto Time Series library project\\Microsoft_Stock.csv')
```

```python
df
```

```python
df=df[['Date','Close']]
df['Date']=pd.to_datetime(df.Date)
df['Close'].plot(figsize=(12,8),title='Microsoft Stock Prices',fontsize=20,label='Close Price')
plt.legend()
plt.grid()
plt.show()
```

```python
from autots import AutoTS
model=AutoTS(forecast_length=10,frequency='infer',ensemble='simple',drop_data_older_than_periods=200)
model=model.fit(df,date_col='Date',value_col='Close',id_col=None)
```

```python
prediction=model.predict()
forecast=prediction.forecast
print("Microsoft Stock Price Prediction for next 10 days")
print(forecast)
```
